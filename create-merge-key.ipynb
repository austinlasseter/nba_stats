```python
import pandas as pd
```

## Read in the datafile

```python
df1 = pd.read_csv('2017-18_teamBoxScore.csv')
```

```python
df2 = pd.read_excel('NBA ODDS 2017-2018.xls')
```

## Create a string variable for date

```python
# What are the data types?
print(df1['gmDate'].dtype)
print(df2['DATE'].dtype)
```

```python
# Create a new string var
df2['newdate']=df2['DATE'].apply(lambda row: row.strftime('%Y-%m-%d'))
df2[['newdate', 'DATE']].head()
```

## Add team abbreviations to the second datafile

```python
# Sean already created these abbreviations, so let's read them in.
df3 = pd.read_csv('NBA All edit.csv')
```

```python
# Extract only the names and abbrevs
abbvs=df3.groupby(['teamAbbr_x', 'Team']).max()
list(abbvs.index)[:5]
```

```python
# COnvert that into a dictionary
abbvs_dict={}
for element in list(abbvs.index):
    abbvs_dict[element[1]]=element[0]
abbvs_dict
```

```python
# Use the dictionary as a map function
df2['teamAbbr']=df2['Team'].map(abbvs_dict)
```

```python
# Confirmt that did what we want
df2[['teamAbbr', 'Team']].head()
```

## Create new columns for team A and team B

```python
# Team A
a=list(df2.loc[::2, 'teamAbbr'])
team_a=[val for val in a for _ in (0, 1)]
df_team_a=pd.DataFrame(team_a, columns=['team_a'])
```

```python
# Team B
b=df2.loc[1::2, 'teamAbbr']
team_b=[val for val in b for _ in (0, 1)]
df_team_b=pd.DataFrame(team_b, columns=['team_b'])
```

```python
# Join them together
df4=pd.concat([df2, df_team_a, df_team_b], axis=1)
```

```python
# Confirm that they're all the same length, which is good.
print(df1.shape)
print(df2.shape)
print(len(team_a))
print(len(team_b))
print(df4.shape)
```

```python
# Take a look
df4.head()
```

## Create a unique game identifier as a merge key

```python
df4['key']=df4['newdate']+'-'+df4['team_a']+'-'+df4['team_b']
```
